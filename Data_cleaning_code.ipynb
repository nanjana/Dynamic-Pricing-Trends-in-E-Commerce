```python
import pandas as pd
import random

# List of known brands
known_brands = ['xiaomi', 'apple', 'huawei', 'acme', 'samsung', 'oppo', 'hp',
       'steelseries', 'oneplus', 'irbis', 'vivo', 'nokia', 'meizu',
       'lenovo', 'honor', 'tp-link', 'elari', 'zte', 'harper', 'sony',
       'acer', 'lg', 'usams', 'bq', 'htc', 'blackberry', 'jbl', 'hyperx',
       'vertex', 'google', 'sennheiser', 'bose', 'huion', 'wacom',
       'kingston', 'prestigio', 'parrot', 'haier', 'razer', 'inoi',
       'jinga', 'doogee', 'nubia', 'creative', 'defender', 'beats',
       'texet', 'marshall', 'panasonic', 'olmio', 'microsoft', 'asus',
       'urbanears', 'awei', 'umi', 'fly', 'nexttab', 'jabra', 'koss',
       'olto', 'edifier', 'plantronics', 'beyerdynamic', 'tecno',
       'canyon', 'philips', 'trust', 'ritmix', 'genius', 'zalman', 'sven',
       'microlab', 'gionee', 'cnd', 'oukitel', 'orico', 'denn',
       'redragon', 'crown', 'audio-technica', 'krk', 'continent',
       'fender', 'wuw', 'leeco', 'inkax', 'micromax', 'xp-pen', 'pioneer',
       'presonus', 'hoco', 'global', 'marley', 'manhattan', 'fiio',
       'veho', 'bluedio', 'fostex', 'logitech', 'alctron', 'ipipoo',
       'qcyber', 'havit', 'leagoo', 'akg', 'ergo', 'joyroom', 'navitel',
       'jvc', 'yamaha', 'gamemax', 'shure', 'hercules', 'rapoo', 'roland',
       'intro', 'westone', 'rombica', 'nuforce', 'monster', 'vox',
       'magnat', 'soundking', 'gigabyte', 'remax', 'cowon', 'elenberg',
       'alesis', 'behringer', 'numark', 'hiper', 'acqua', 'superlux',
       'delux', 'ultrasone', 'msi', 'tascam', 'hifiman', 'dunu', 'meze',
       'g-cube', 'ark', 'samson', 'nomi', 'proel', 'siee', 'casio',
       'm-audio', 'tesoro', 'avermedia', 'celebrat', 'skullcandy',
       'gembird', 'sades', 'kalenji', 'borasco', 'nobby', 'gamdias',
       'motorola', 'cougar', 'aula', 'bravis', 'yison', 'mark', 'iwalk',
       'invotone', 'moxom'
]

# Task 1: Extract brand name from title if found in the list, otherwise assign "Unknown Brand"
def extract_brand(title):
    if not isinstance(title, str):
        return "Unknown Brand"
    for brand in known_brands:
        if brand.lower() in title.lower():  # Case-insensitive search
            return brand
    return "Unknown Brand"

# Apply the function to populate the 'brand' column
df_ebay_filter['brand'] = df_ebay_filter['title'].apply(extract_brand)

# Save the updated DataFrame
#df.to_csv('updated_dataset_with_brands.csv', index=False)

print("Brand extraction complete!")
```

```python
# Task 2: Populate the 'shipping_price' column
def populate_shipping_price():
    # Randomly assign 0 or a shipping price between 1 to 15
    return 0 if random.random() < 0.3 else random.randint(1, 15)  # 30% chance for 0, 70% for 1-15

# Apply the function to populate the 'shipping_price' column
df_ebay_filter['shipping_price'] = df_ebay_filter['shipping_price'].apply(lambda x: populate_shipping_price())
```

```python
# Task 2: Populate the 'ratings' column
def populate_ratings(existing_value):
    if pd.isnull(existing_value):  # If the value is NaN, assign a random rating
        possible_ratings = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
        return random.choice(possible_ratings)
    return existing_value  # Leave existing ratings as they are

# Apply the function to populate the 'ratings' column
df_ebay_filter['ratings'] = df_ebay_filter['ratings'].apply(populate_ratings)
```
